```python
#연설문 불러오기
moon = open('speech_moon.txt', encoding = 'UTF-8').read()
moon
```

```python
#불필요한 문자 제거하기
import re
moon = re.sub('[^가-힣]', ' ', moon)
moon
```

```python
#명사 추출하기
import konlpy
hannanum = konlpy.tag.Hannanum()

hannanum.nouns("대한민국의 영토는 한반도와 그 부속도서로 한다")
```

```python
nouns = hannanum.nouns(moon)
nouns
```

```python
import pandas as pd
df_word = pd.DataFrame({'word' : nouns})
df_word
```

```python
#글자 수 추가
df_word['count'] = df_word['word'].str.len()
df_word
```

```python
#두 글자 이상 단어만 남기기
df_word = df_word.query('count >= 2')
df_word.sort_values('count')
```

```python
#단어 빈도 구하기
df_word = df_word.groupby('word', as_index = False)\
.agg(n = ('word', 'count'))\
.sort_values('n', ascending = False)

df_word
```

```python
#단어 빈도 상위 20개 추출
top20 = df_word.head(20)
top20
```

```python
#데이터 프레임을 딕셔너리로 변환
dic_word = df_word.set_index('word').to_dict()['n']
dic_word
```

```python
#워드 클라우드 만들기
img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize = (10, 10)) #가로, 세로 설정
plt.axis('off') #테두리 선 없애기
plt.imshow(img_wordcloud) #워드 클라우드 출력
```

#10-2. 기사 댓글 텍스트 마이닝

```python
#기사 댓글 불러오기
df = pd.read_csv('news_comment_BTS.csv', encoding = 'UTF-8')
df.info()
```

```python
#불필요한 문자 제거하기
#moon같은 경우는 str이기 때문에 info()에서 에러가 남
#이 기사의 경우는 데이터 프레임에 담겨있는 변수이므로 str.replace사용
df['reply'] = df['reply'].str.replace('[^가-힣]', ' ', regex = True)
df['reply'].head()
```

```python
import konlpy
kkma = konlpy.tag.Kkma() #Kkma -> 꼬꼬마 형태 분석기
```

```python
#명사 추출 - apply() 활용
nouns = df['reply'].apply(kkma.nouns)
nouns
```

```python
#한 행에 한 단어가 들어가도록 구성
nouns = nouns.explode()
nouns
```

```python
#데이터 프레임 만들기
df_word = pd.DataFrame({'word' : nouns})

#글자 수 추가
df_word['count'] = df_word['word'].str.len()

#두 글자 이상 단어만 남기기
df_word = df_word.query('count >= 2')
df_word
```

```python
#빈도표 만들기
df_word = df_word.groupby('word', as_index =False)\
.agg(n = ('word', 'count'))\
.sort_values('n', ascending = False)

df_word
```

```python
#top20 추출
top20 = df_word.head(20)
top20
```

```python
plt.rcParams.update({'figure.figsize' : [6.5, 6]})
sns.barplot(data = top20, y = 'word', x = 'n')
```

```python
dic_word = df_word.set_index('word').to_dict()['n']
```

```python
wc = WordCloud(random_state = 1234,
               font_path = font,
               width = 400,
               height = 400,
               background_color = 'white',
               mask = img)
```

```python
img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize = (10, 10))
plt.axis('off')
plt.imshow(img_wordcloud)
```
