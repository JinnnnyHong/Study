# 6-6 집단별로 요약하기

```python
# 집단별로 요약하기
exam.agg(mean_math = ('math', 'mean'))
```

```python
exam.groupby('nclass')\
.agg(mean_math = ('math','mean'))
```

```python
exam.groupby('nclass')\
.agg(mean_math = ('math', 'mean'),
     sum_math = ('math', 'sum'),
     median_math = ('math', 'median'),
     n = ('nclass', 'count')) #학생수
```

```python
exam.groupby('nclass').mean()
```

```python
# 집단별로 다시 집단 나누기
mpg.groupby(['manufacturer', 'drv'])\ #제조회사 및 구동 방식별 분리
.agg(mean_cty = ('cty', 'mean')) # cty 평균 구하기
```

```python
#audi의 drv별 빈도
mpg.query('manufacturer == "audi"')\
.groupby(['drv'])\
.agg(n = ('drv', 'count'))
```

```python
#chevrolet의 drv별 빈도
mpg.query('manufacturer == "chevrolet"')\
.groupby(['drv'])\
.agg(n = ('drv', 'count'))
```

```python
mpg.groupby('drv')\
.agg(n = ('drv', 'count'))
```

```python
mpg['drv'].value_counts()
```

```python
#pandas 합수 조합하기
# Q. 제조 회사별로 'suv' 자동차의 도시 및 고속도로 합산 연비 평균을 구해 내림차순으로 정렬하고, 1~5위까지 출력
mpg.query('category == "suv"')\
.assign(total = (mpg['hwy']+mpg['cty'])/2)\
.groupby('manufacturer')\
.agg(mean_tot = ('total', 'mean'))\
.sort_values('mean_tot', ascending = False)\
.head(5)
```

# 6-7 데이터 합치기

```python
#가로로 합치기
test1 = pd.DataFrame({'id': ['1', '2', '3', '4', '5'],
                     'midterm' : ['60','80','70','90','85']})

test2 = pd.DataFrame({'id': ['1', '2', '3', '4', '5'],
                     'final' : ['70','83','65','95','80']})
```

```python
total = pd.merge(test1, test2, how = 'left', on = 'id')
total
```

```python
name = pd.DataFrame({'nclass': [1,2,3,4,5],
                     'teacher': ['kim', 'lee','park','choi', 'jung']})
name
```

```python
exam_new = pd.merge(exam, name, how = 'left', on = 'nclass')
exam_new
```

```python
#세로로 합치기
total2 = pd.concat([test1, test2])
total2
```
