```python
mpg['total'] = (mpg['cty']+mpg['hwy'])/2
mpg.head()
```

```python
mpg['total'].mean()
```

```python
mpg['total'].describe()
```

```python
mpg['total'].plot.hist()
```

```python
import numpy as np
```

```python
mpg['test'] = np.where(mpg['total']>=20, 'pass', 'fail')
mpg.head()
```

```python
mpg['test'].value_counts()
```

```python
count_test = mpg['test'].value_counts() #빈도표를 변수화
```

```python
count_test.plot.bar(rot = 0) # rot=0 축 이름 수평 만들기, 아무것도 설정 안할 시 이름이 90도로 누워짐
```

## 중첩 조건문 활용

```python
mpg['grade'] = np.where(mpg['total']>=30, 'A',
                        np.where(mpg['total']>=20, 'B', 'C'))
mpg.head()
```

```python
count_grade = mpg['grade'].value_counts()
count_grade
```

```python
count_grade.plot.bar(rot = 0)
```
